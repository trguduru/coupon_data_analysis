### Data Cleanup
In this notebook we look at the data provided and see if anything is missing, have nulls values, outliers etc ... and then do some cleanup to make sure the data is in a usable state for further analysis

##### [NOTE:]The comments on the analysis is documented after the each cell code

```python
# import all required libraries for data cleanup
import pandas as pd
import numpy as np
```

```python
# load the coupons.csv data using read_csv method in pandas

data = pd.read_csv('data/coupons.csv')
```

```python
#inspect to see whether the dataframe has any data or not
data.head()
```

```python
# check how many non-null counts for each feature
data.info()
```

```python
# find out the number of null values for each given column
data.isna().sum()
```

In above sum of null values, are many columns has nulls value, to name them car,bar,CoffeeHouse, CarryAway,RestaurantLessThan20 and Restaurant20To50.

Some of these features(columns) are very important to answer the question of 'will customer accepts the coupon' for example bar,CoffeeHouse, CarryAway,RestaurantLessThan20 and Restaurant20To50.

So we may have to replace these nulls with maybe a median of the data if the column is of numerical data else we may have to drop the column data.


```python
# Lets drop the null values
data.dropna()
```

The above dropna() mwthod will drop most of the rows as 'car'has more rows with null values.
The resultant dataframe has only 108 which far less than original dataset which 12684, so dropping null value columns is not a good idea.

```python
# Lets check how categorical the 'car' column is and what values it has
data['car'].value_counts()
```

When we look at the 'car' column values i dont think it will add any value for this analysis so either we can drop just the column of leave it as is and dont use this column for analysis purpose.

For the purpose of this study im going to drop the column for now and see how it impacts the study.

```python
# dropping the column 'car'
data.drop('car', axis=1, inplace=True)
```

After looking at the values of the above columns, the values **less1/never** goes with value as **'0'** as these are visits and can only be whole number. So we can replace **'less1'** and **'never'** with **'0/never'** similar to other categorical values.

Updating these column values as **'never'**

```python
# replace the all column values of less1 to never
data['Bar'].mask(data['Bar'] == 'less1','never', inplace=True)
data['CoffeeHouse'].mask(data['CoffeeHouse'] == 'less1','never', inplace=True)
data['CarryAway'].mask(data['CarryAway'] == 'less1','never', inplace=True)
data['RestaurantLessThan20'].mask(data['RestaurantLessThan20'] == 'less1','never', inplace=True)
data['Restaurant20To50'].mask(data['Restaurant20To50'] == 'less1','never', inplace=True)
```

```python
# lets check to make sure the values are updated
data['CoffeeHouse'].value_counts()
```

### Fill null values
The null value columns are bar,CoffeeHouse, CarryAway,RestaurantLessThan20 and Restaurant20To50. We can fill these columns with 'never' as the null values are very less and 'never' is the value of the most of the rows.

```python
# fill all the above mentioned column values to 'never'
data.fillna('never', inplace=True)
```

```python
# now no null values in the data set
data.info()
```

### Column Names
Lets look at the column names and see if they are makes sense or not. If not rename them to easy to understand.

The last column name is **'Y'** which stands for whether the coupon is accepted or not. Lets rename this to **'coupon_accepted'** to easy to read.

```python
# rename the column 'Y'
data.rename({'Y':'coupon_accepted'}, axis='columns', inplace=True)
```

```python
# check to make sure whether its renamed or not
data.info()
```

```python
#Find the dupliates of the dataset
data[data.duplicated()]
```

```python
#Remove duplicate rows
data.drop_duplicates(keep='first', inplace=True)
```

```python
# Save the processing data as 'coupons_processed.csv'
data.to_csv('data/coupons_processed.csv',index=False)
```
